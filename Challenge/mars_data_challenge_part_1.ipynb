```python
# Scraping table with Splinter and Beautiful Soup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import json

# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# Visit the Mars Facts site
url = 'https://redplanetscience.com'
browser.visit(url)

# Assigned HTML content to the html variable
html = browser.html
# Beautiful Soup then parses the HTML content.
html_soup = soup(html, 'html.parser')
```

```python
# articles = html_soup.find_all('div', class_='list_text')
articles = html_soup.find('div', id='news', class_='container')
articles
```

```python
list_text = articles.find_all('div', class_='list_text')
list_text
```

```python
news_list = []
```

```python
for article in list_text:
    title = article.find('div', class_='content_title').text
    preview = article.find('div', class_='article_teaser_body').text
    news_dict = {}
    news_dict["title"] = title
    news_dict["preview"] = preview
    news_list.append(news_dict)
news_list
```

```python
# Store the scraped data in a JSON file 
print(json.dumps(news_list))
with open("outfile.json", "w") as f:
    f.write(json.dumps(news_list))
```

```python
# End the Automated Browsing Session
browser.quit()
```
